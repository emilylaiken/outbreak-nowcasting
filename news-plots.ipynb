```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def scale(l):
    d = max(l)
    return [e/d for e in l]
```

```python
fig, axs = plt.subplots(5, 1,figsize=(14,10))
```

```python
# Yemen Cholera.
df_predictors = pd.read_csv('./data/yemen/yemenpredictors.csv')
df_reports = pd.read_csv('./data/yemen/yemenreports.csv')
df_gdelt = pd.read_csv('./gdelt_data/cholera_yemen_A.csv',header=None)

#print(df_predictors.columns)

df_predictors['date'] = df_predictors['date'].astype('datetime64[ns]')
df_reports['date'] = df_reports['date'].astype('datetime64[ns]')
df_gdelt['date'] =  (df_gdelt[1] + '/' + df_gdelt[0].astype(str)).astype('datetime64[ns]')
df_gdelt['news'] = df_gdelt[2]

term = 'cholera'
#plt.figure(figsize=(10,5))
axs[0].plot(df_predictors['date'],scale(df_predictors[term]),label='Search Volume',c='r',alpha=0.7)
axs[0].plot(df_reports['date'],scale(df_reports['groundtruth']),label='Case Counts',c='k',alpha=0.7)
axs[0].plot(df_gdelt['date'],scale(df_gdelt['news']),label='News',c='b',alpha=0.7)
axs[0].legend()
axs[0].set_title('Cholera - Yemen')
# plt.xlabel('Date')
axs[0].set_ylabel('Scaled Volume')
# plt.savefig('news_cholera_yemen.svg', format='svg')
```
